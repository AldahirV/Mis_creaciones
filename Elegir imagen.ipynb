```python
import cv2

# Leer las imágenes y convertirlas a RGB
def read_image(path):
    image = cv2.imread(path)
    if image is None:
        print(f"Error: No se pudo leer la imagen en {path}")
        return None
    return cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

perro_1 = read_image('C:/Users/admin/Documents/ALDAHIR FELIZ/Proyectos mios/Proyecto_1/perro_realista.jpg')
perro_2 = read_image('C:/Users/admin/Documents/ALDAHIR FELIZ/Proyectos mios/Proyecto_1/perro_animado.webp')
robot_1 = read_image('C:/Users/admin/Documents/ALDAHIR FELIZ/Proyectos mios/Proyecto_1/robot_realista.jpg')
robot_2 = read_image('C:/Users/admin/Documents/ALDAHIR FELIZ/Proyectos mios/Proyecto_1/robot_animado.jpg')
casa_1 = read_image('C:/Users/admin/Documents/ALDAHIR FELIZ/Proyectos mios/Proyecto_1/casa_realista.jpg')
casa_2 = read_image('C:/Users/admin/Documents/ALDAHIR FELIZ/Proyectos mios/Proyecto_1/casa_animada.jpg')

# Función para mostrar la imagen en una nueva ventana
def show_image(title, image):
    if image is not None:
        # Convertir de RGB a BGR para mostrar con OpenCV
        image_bgr = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)
        cv2.imshow(title, image_bgr)
        cv2.waitKey(0)
        cv2.destroyAllWindows()
    else:
        print(f"Error: La imagen no se pudo mostrar.")

# Antes de elegir la imagen
print("Hola, usted va a elegir una imagen y yo se la voy a mostrar.")
print("Elija una categoría: \n 1) Perro \n 2) Robot \n 3) Casa")
eleccion = input("Introduzca el número de la categoría que desea: ")

# Condiciones para elegir la imagen
if eleccion == "1":
    print("Elija: \n 1) Perro realista \n 2) Perro animado")
    AWSD_1 = input("Introduzca el número de la imagen que desea: ")
    if AWSD_1 == "1":
        show_image('Perro Realista', perro_1)
    elif AWSD_1 == "2":
        show_image('Perro Animado', perro_2)
    else:
        print("No ha introducido un número válido.")

elif eleccion == "2":
    print("Elija: \n 1) Robot realista \n 2) Robot animado")
    AWSD_2 = input("Introduzca el número de la imagen que desea: ")
    if AWSD_2 == "1":
        show_image('Robot Realista', robot_1)
    elif AWSD_2 == "2":
        show_image('Robot Animado', robot_2)
    else:
        print("No ha introducido un número válido.")

elif eleccion == "3":
    print("Elija: \n 1) Casa realista \n 2) Casa animada")
    AWSD_3 = input("Introduzca el número de la imagen que desea: ")
    if AWSD_3 == "1":
        show_image('Casa Realista', casa_1)
    elif AWSD_3 == "2":
        show_image('Casa Animada', casa_2)
    else:
        print("No ha introducido un número válido.")
else:
    print("No ha introducido un número válido.")
```
